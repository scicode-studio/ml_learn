```python

def printSurvivalRate(gender):
    all = titanic_train.loc[titanic_train.Sex == gender]["Survived"]
    num_all = all.shape[0]
    num_survived = sum(all)

    print("All {0} count: {1}, survived: {2}, ratio: {3}%".format(gender, num_all, num_survived, num_survived / num_all * 100) )

printSurvivalRate("male")
printSurvivalRate("female")
```
